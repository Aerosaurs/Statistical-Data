```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('ecommerce_customer_data.csv')

# Set plot size
plt.figure(figsize=(8, 6))

# Distribution of Age
sns.histplot(data['Age'], kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('ecommerce_customer_data.csv')

# Set plot size
plt.figure(figsize=(8, 6))

# Boxplot of Income by Gender
sns.boxplot(x='Gender', y='Income', data=data, palette='Set2')
plt.title('Income Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Income')

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('ecommerce_customer_data.csv')

# Set plot size
plt.figure(figsize=(8, 6))

# Barplot of Product Category by Gender
sns.countplot(x='Product_Category', hue='Gender', data=data, palette='husl')
plt.title('Product Category Count by Gender')
plt.xlabel('Product Category')
plt.ylabel('Count')

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('ecommerce_customer_data.csv')

# Set plot size
plt.figure(figsize=(8, 6))

# Scatterplot of Income vs Total Spend
sns.scatterplot(x='Income', y='Total_Spend', hue='Gender', data=data, palette='coolwarm')
plt.title('Income vs Total Spend')
plt.xlabel('Income')
plt.ylabel('Total Spend')

# Show plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('ecommerce_customer_data.csv')

# Check column names
print(data.columns)

# Ensure column names are clean
data.columns = data.columns.str.strip()

# Pairplot to explore relationships
sns.pairplot(data[['Age', 'Income', 'Total_Spend', 'Num_Purchases', 'Product_Category']], hue='Product_Category', palette='viridis')
plt.suptitle('Pairplot of Age, Income, Total Spend, and Purchases', y=1.02)

# Show plot
plt.show()
```
